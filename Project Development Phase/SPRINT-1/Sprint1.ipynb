```python

import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_excel("/Global_Superstore2.xlsx")
```

```python
plt.bar(data['Segment'], data['Sales'])
 
plt.title("Segment Wise Sales")
 
# Setting the X and Y labels
plt.xlabel('Segment')
plt.ylabel( 'Sales')
 
# Adding the legends
plt.show()
```

```python
plt.bar(data['Segment'], data['Profit'])
 
plt.title("Segment Wise Profit")
 
# Setting the X and Y labels
plt.xlabel('Segment')
plt.ylabel( 'Profit')
 
# Adding the legends
plt.show()
```

```python
plt.bar(data['Segment'], data['Quantity'])
 
plt.title("Segment Wise Quantity")
 
# Setting the X and Y labels
plt.xlabel('Segment')
plt.ylabel( 'Quantity')
 
# Adding the legends
plt.show()
```
